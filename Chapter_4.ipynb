# BA4B : Find Substrings of a Genome Encoding a Given Amino Acid String

```python
inp = "ATGGCCATGGCCCCCAGAACTGAGATCAATAGTACCCGTATTAACGGGTGA"
amino = "MA"
l = len(inp)
aminol = len(amino)
k = 3*aminol

def checking(s):
    val = 3
    ck = 1
    for j in range(0,aminol):
        p = s[val*j: val*j+3]
        try:
            if(dic[p] != amino[j]):
                ck = 0
                break
        except:
            ck = 0
    return ck

for i in range(0,l-k+1):
    s = inp[i:i+k]
    ###Normal part
    if checking(s) == 1:
        print(s)
    s1 = ""
    ###Transcription
    for j in range(0,k):
        if s[j] == "T":
            s1 += "U"
        else:
            s1 += s[j]
    if checking(s1) == 1:
        print(s)
     
    s1 = ""
    ###Reverse comp. with 'U'
    for j in range(0,k):
        if s[j] == "T":
            s1 += "A"
        elif s[j] == "A":
            s1 += "U"
        elif s[j] == "G":
            s1 += "C"
        elif s[j] == "C":
            s1 += "G"
    s1 = "".join(reversed(s1))
    if checking(s1) == 1:
        print(s)
```

# BA4E : Find a Cyclic Peptide with Theoretical Spectrum Matching an Ideal Spectrum

```python
st = "LEQN"
spectrum = [0, 113, 128, 186, 241, 299, 314, 427]
mp = {}
for i in range(0,len(spectrum)):
    mp[spectrum[i]] = 1
v = len(spectrum)
valu = 427
ss = st+st
d = {}
def banaicyclic(st):
    ss = st+st
    l = {}
    for i in range(1,len(st)):
        for j in range(0,len(st)):
            sss = ""
            for k in range(j,j+i):
                sss = sss+ss[k]
            l[sss] = 1
    l[st] = 1
    return l

def banailinear(st):
    l = {}
    for i in range(1,len(st)):
        for j in range(0,len(st)-i+1):
            sss = ""
            for k in range(j,j+i):
                sss = sss+st[k]
            l[sss] = 1
    l[st] = 1
    return l

def val(a):
    if(a=='G'):
        return 57
    if(a=='A'):
        return 71
    if(a=='S'):
        return 87
    if(a=='P'):
        return 97
    if(a=='V'):
        return 99
    if(a=='T'):
        return 101
    if(a=='C'):
        return 103
    if(a=='I'):
        return 113
    if(a=='N'):
        return 114
    if(a=='D'):
        return 115
    if(a=='Q'):
        return 128
    if(a=='E'):
        return 129
    if(a=='M'):
        return 131
    if(a=='H'):
        return 137
    if(a=='F'):
        return 147
    if(a=='R'):
        return 156
    if(a=='Y'):
        return 163
    if(a=='W'):
        return 186
    
    
def fun(x):
    valu = 0
    for i in x:
        valu = valu + val(i)
    return valu
ans = []
amin = ['G','A','S','P','V','T','C','I','N','D','Q','E','M','H','F','R','Y','W']

def sequence(valu):
    peptides =[]
    peptides.append("")
    while(1):
        g = len(peptides)
        if g == 0:
            break
        pep = []
        for i in range(0,len(peptides)):
            for j in range(0,len(amin)):
                pep.append(peptides[i]+amin[j])
        del peptides[:]
        for i in range(0,len(pep)):
            peptides.append(pep[i])
        del pep[:]
        for i in range(0,len(peptides)):
            x = fun(peptides[i])
            y = peptides[i]
            if x == valu:
                z = banaicyclic(y)
                t = []
                for j in z:
                    t.append(fun(j))
                t.append(0)
                t.sort()
                flag = 1
                if(len(t) != len(spectrum)):
                    flag = 0
                if flag == 1:
                    for k in range(0,len(t)):
                        if t[k] != spectrum[k]:
                            flag = 0
                            break
                if flag == 1:
                    ans.append(y)
            else:
                z = banailinear(y)
                t = []
                for j in z:
                    t.append(fun(j))
                t.append(0)
                t.sort()
                flag = 1
                for k in range(0, len(t)):
                    if t[k] in mp:
                        continue
                    else:
                        flag = 0
                        break
                if flag == 1:
                    pep.append(y)
        
        del peptides[:]
        for i in range(0,len(pep)):
            peptides.append(pep[i])
        del pep[:] 
sequence(valu)

for i in range(0,len(ans)):
    x = ans[i]
    cnt = 1
    ll = len(x)
    for k in x:
        print(val(k),end = "")
        if(cnt==ll):
            print(" ",end = "")
        else:
            print("-",end = "")
        cnt = cnt+1
    
```
